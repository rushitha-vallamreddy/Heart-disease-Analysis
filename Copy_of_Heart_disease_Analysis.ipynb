```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df.target.value_counts()
```

```python
corr=df.corr()
plt.figure(figsize=(18,10)) # Now you can call functions from the pyplot module using plt
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['Not Disesase','Disease'])
plt.title("Target Distribution");

```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA538','red'])
plt.legend(["Haven't Disease ,Have Disease"])
plt.title("heart disease frequency for ages")
plt.xlabel("age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='green',label='Not Disease')
sns.distplot(df.age[df['target']==1],bins=30,color='red',label="Disease")
plt.legend()
plt.title("Heart disease  distribution for ages")
plt.xlabel('age')
plt.ylabel('frquency')
plt.show()
```
